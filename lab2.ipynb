```python
import numpy as np
import matplotlib.pyplot as plt

temps_celsius = np.array([12, 15, 20, 25, 30, 35, 33, 32, 28, 22, 18, 14])
months = np.array(["Jan", "Feb", "Mar", "Apr", "May", "Jun",
                   "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])

print("Monthly Temperatures (째C):", temps_celsius)

avg_temp = np.mean(temps_celsius)
hottest_month = months[np.argmax(temps_celsius)]
coldest_month = months[np.argmin(temps_celsius)]

print("\nAverage Temperature (째C):", avg_temp)
print("Hottest Month:", hottest_month, "with", temps_celsius.max(), "째C")
print("Coldest Month:", coldest_month, "with", temps_celsius.min(), "째C")

normalized_temps = (temps_celsius - temps_celsius.min()) / (temps_celsius.max() - temps_celsius.min())
print("\nNormalized Temperatures:", normalized_temps)

above_avg_months = months[temps_celsius > avg_temp]
print("\nMonths above annual average:", above_avg_months)

temps_fahrenheit = temps_celsius * 9/5 + 32
print("\nTemperatures in Fahrenheit:", temps_fahrenheit)

plt.plot(months, temps_celsius, marker='o', label="Celsius")
plt.plot(months, temps_fahrenheit, marker='s', label="Fahrenheit")
plt.axhline(avg_temp, color='r', linestyle="--", label=f"Avg Temp {avg_temp:.2f}째C")
plt.title("Monthly Temperature Trend")
plt.xlabel("Month")
plt.ylabel("Temperature")
plt.legend()
plt.grid(True)
plt.show()
```
